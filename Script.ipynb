```python
#Import libraries
import pandas as pd
import tensorflow as tf
```

```python
#Load data
Traindata = pd.read_csv('Data/train.csv', header=None)
Vocab = pd.read_csv('Data/vocab.csv', header=None)
Intents = pd.read_csv('Data/intents.csv', header=None)
Trainintents = pd.read_csv('Data/train_intent.csv', header=None)
Testdata = pd.read_csv('Data/test.csv', header=None)
Slots = pd.read_csv('Data/slots.csv', header=None)
Trainslots = pd.read_csv('Data/train_slots.csv', header=None)
```

```python
#create sentence
sentences = []
for line in Traindata[0]:
    sentence = []
    for wordid in line.split():
        sentence.append(Vocab[0][int(wordid)])
    sentence = ' '.join(sentence)
    print(sentence)
    sentences.append(sentence)
```

```python
#create intents
intents = []
for intentid in Trainintents[0]:
    intents.append(Intents[0][int(intentid)])
```

```python
#create slots
slots = []
for line in Trainslots[0]:
    slot = []
    for slotid in line.split():
        slot.append(Slots[0][int(slotid)])
    print(slot)
    slot = ' '.join(slot)
    slots.append(slot)
```

```python
#Make datasets
dataintents = pd.DataFrame(data = list(map(list, zip(*[sentences, intents]))), columns = ['Sentence', 'Intent'])
dataslots = pd.DataFrame(data = list(map(list, zip(*[sentences, slots]))), columns = ['Sentence', 'Slots'])
```

```python
#visualize using tensorboard
dataintenttensor = tf.convert_to_tensor(value = dataintents.values[1:10])
dataslottensor = tf.convert_to_tensor(value = dataslots.values[1:10])
tf.summary.text('DataIntent', dataintenttensor)
tf.summary.text('DataSlots', dataslottensor)
merged = tf.summary.merge_all()
with tf.Session() as sess:
    summary = sess.run(merged)
    data_writer = tf.summary.FileWriter('./tb')
    data_writer.add_summary(summary, 0)
```

```python
nltk.download("punkt") #pretrained tokenizer
nltk.download("stopwords") #words like and, or, of etc., which do not have semantic meaning
```

```python
file_name = "train_sentences.txt"
corpus_raw = u""
with codecs.open(file_name, "r", "utf-8") as book_file:
    corpus_raw += book_file.read()
print("Corpus is now {0} characters long".format(len(corpus_raw)))
print()
```

#Tokenization 

```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
```

```python
raw_sentences = tokenizer.tokenize(corpus_raw)
```

```python
def sentence_to_wordlist(raw):
    clean = re.sub("[^a-zA-Z]"," ", raw)
    words = clean.split()
    return words
```

```python
tokenized_sentences = []
for raw_sentence in raw_sentences:
    if len(raw_sentence) > 0:
        tokenized_sentences.append(sentence_to_wordlist(raw_sentence))
```

```python
#print(raw_sentences[0])
print(sentence_to_wordlist(raw_sentences[0]))
```

```python
token_count = sum([len(sentence) for sentence in sentences])
print("The book corpus contains {0:,} tokens".format(token_count))
```

# word2vec

```python
num_features = 300
min_word_count = 3
num_workers = multiprocessing.cpu_count()
context_size = 7
downsampling = 1e-3
seed = 1
```

```python
vectors = w2v.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample=downsampling
)
```

```python
vectors.build_vocab(tokenized_sentences)
```

```python
print("Word2Vec vocabulary length:", len(vectors.wv.vocab))
```

```python
vectors.train(tokenized_sentences,total_examples=vectors.corpus_count, epochs=vectors.iter)
```

```python
if not os.path.exists("trained"):
    os.makedirs("trained")
```

```python
vectors.save(os.path.join("trained", "vectors.bin"))
```

```python
vectors = w2v.Word2Vec.load(os.path.join("trained", "vectors.bin"))
```

```python
tsne = sklearn.manifold.TSNE(n_components=2, random_state=0) #compress word vectors to 2D
all_word_vectors_matrix = vectors.wv.syn0
all_word_vectors_matrix_2d = tsne.fit_transform(all_word_vectors_matrix)
points = pd.DataFrame(
    [
        (word, coords[0], coords[1])
        for word, coords in [
            (word, all_word_vectors_matrix_2d[vectors.wv.vocab[word].index])
            for word in vectors.wv.vocab
        ]
    ],
    columns=["word", "x", "y"]
)
points.head(100)
sns.set_context("poster")
points.plot.scatter("x", "y", s=10, figsize=(20, 12))
def plot_region(x_bounds, y_bounds):
    slice = points[
        (x_bounds[0] <= points.x) &
        (points.x <= x_bounds[1]) & 
        (y_bounds[0] <= points.y) &
        (points.y <= y_bounds[1])
    ]
    
    ax = slice.plot.scatter("x", "y", s=35, figsize=(10, 8))
    for i, point in slice.iterrows():
        ax.text(point.x + 0.005, point.y + 0.005, point.word, fontsize=11)
plot_region(x_bounds=(0.0, 10.0), y_bounds=(0.0, 10.0))
vectors.most_similar("boeing")
```
