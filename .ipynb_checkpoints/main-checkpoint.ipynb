```python
#Import libraries
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load data
Traindata = pd.read_csv('Data1/train.csv', header=None)
Vocab = pd.read_csv('Data1/vocab.csv', header=None)
Intents = pd.read_csv('Data1/intents.csv', header=None)
Trainintents = pd.read_csv('Data1/train_intent.csv', header=None)
Testdata = pd.read_csv('Data1/test.csv', header=None)
Slots = pd.read_csv('Data1/slots.csv', header=None)
Trainslots = pd.read_csv('Data1/train_slots.csv', header=None)
```

```python
#create sentence for train data
sentences = []
for line in Traindata[0]:
    sentence = []
    for wordid in line.split():
        sentence.append(Vocab[0][int(wordid)])
    print(sentence)
    sentences.append(sentence)
```

```python
#create sentences for test data
test_sentences = []
for line in Testdata[0]:
    test_sentence = []
    for wordid in line.split():
        test_sentence.append(Vocab[0][int(wordid)])
    #sentence = ' '.join(sentence)
    print(test_sentence)
    test_sentences.append(test_sentence)
```

```python
#create intents
intents = []
for intentid in Trainintents[0]:
    intents.append(Intents[0][int(intentid)])
print(intents)
```

```python
#create slots
slots = []
for line in Trainslots[0]:
    slot = []
    for slotid in line.split():
        slot.append(Slots[0][int(slotid)])
    #slot = ' '.join(slot)
    print(slot)
    slots.append(slot)
```

```python
#Padding train sentences to get all sentences to fixed length which is 50 here
sentences = [t[:-1] for t in sentences]
length = 50
sin = []
sout = []
max_train_input=0
max_train_slot=0
for line in range(len(sentences)):
    temp = sentences[line]
    if(len(temp)>max_train_input):
        max_train_input=len(temp)
    #print(len(temp))
    if len(temp) < length:
        temp.append('EOS')
        while len(temp) < length:
            temp.append('<PAD>')
    else:
        temp = temp[:length]
        temp.append('EOS')
    sin.append(temp)
        
for line1 in range(len(slots)): 
    temp1 = slots[line1]
    if(len(temp1)>max_train_slot):
        max_train_slot=len(temp1)
    if len(temp1) < length:
        while len(temp1) < length:
            temp1.append('<PAD>')
    else:
        temp1 = temp1[:length]
        #temp1.append('<EOS>')
    sout.append(temp1)
data = list(zip(sin, sout, intents))
print(data[0])
print(max_train_input)
print(max_train_slot)
```

```python
#Padding test sentences to get all sentences to fixed length which and make blank slots and intents for given test sentences 
test_sentences = [t[:-1] for t in test_sentences]
length = 50
test_sin = []
test_sout = []
test_intent = []
max_test_input=0
for line in range(len(test_sentences)):
    temp = test_sentences[line]
    if(len(temp)>max_test_input):
        max_test_input=len(temp)
    if len(temp) < length:
        temp.append('EOS')
        while len(temp) < length:
            temp.append('<PAD>')
    else:
        temp = temp[:length]
        temp.append('EOS')
    test_sin.append(temp)
    count=0
    temp=[]
    while count<length:
        temp.append('<PAD')
        count+=1
    test_sout.append(temp)
    temp=""
    test_intent.append(temp)
test_data = list(zip(test_sin,test_sout,test_intent))
print(test_data[0])
print(max_test_input)
```

```python
#Maps words to indices with the help of vocabulary list
flatten = lambda l: [item for sublist in l for item in sublist]
index_seq2slot = lambda s, index2slot: [index2slot[i] for i in s]
index_seq2word = lambda s, index2word: [index2word[i] for i in s]
seq_in, seq_out, intent = list(zip(*data))
#Vocabulary list for input sentences
vocab = set(flatten(seq_in))
#Vocabulary list for Slots
slot_tag = set(flatten(seq_out))
#Vocabulary list for intents
intent_tag = set(intent)
# generate index for words 
word2index = {'<PAD>': 0, '<UNK>': 1, 'BOS': 2, 'EOS': 3}
for token in vocab:
    if token not in word2index.keys():
        word2index[token] = len(word2index)

# generate words for indices
index2word = {v: k for k, v in word2index.items()}

# generate slots for indices
tag2index = {'<PAD>': 0, '<UNK>': 1, "O": 2}
for tag in slot_tag:
    if tag not in tag2index.keys():
        tag2index[tag] = len(tag2index)

# generate indices for slots
index2tag = {v: k for k, v in tag2index.items()}

# generate indices for intents
intent2index = {'<UNK>': 0}
for ii in intent_tag:
    if ii not in intent2index.keys():
        intent2index[ii] = len(intent2index)

# generate index2intent
index2intent = {v: k for k, v in intent2index.items()}

index2word
```

```python
#Replaces words in the train data with corresponding indices
new_data = []
for sin, sout, intent in data:
    sin_ix = list(map(lambda i: word2index[i] if i in word2index else word2index["<UNK>"],sin))
    true_length = sin.index('EOS')
    sout_ix = list(map(lambda i: tag2index[i] if i in tag2index else tag2index["<UNK>"],sout))
    intent_ix = intent2index[intent] if intent in intent2index else intent2index["<UNK>"]
    new_data.append([sin_ix, true_length, sout_ix, intent_ix])
#Splitting data set into Training set(~70% of data) and Validation set(~30% of data)
train_data=new_data[:3500]
val_data=new_data[3500:]
print('Input_sequence | Actual_length_of_input(Unpadded) | Slot | Intent\n')
print(new_data[0])
print(train_data[0])
print(val_data[0])
```

```python
#Replaces words in the test data with corresponding indices
new_test_data = []
for sin, sout, intent in test_data:
    sin_ix = list(map(lambda i: word2index[i] if i in word2index else word2index["<UNK>"],sin))
    true_length = sin.index('EOS')
    sout_ix = list(map(lambda i: tag2index[i] if i in tag2index else tag2index["<UNK>"],sout))
    intent_ix = intent2index[intent] if intent in intent2index else intent2index["<UNK>"]
    new_test_data.append([sin_ix, true_length, sout_ix, intent_ix])
print('Input_sequence | Actual_length_of_input(Unpadded) | Slot | Intent\n')
print(new_test_data[0])
```

```python
import random

input_steps = 50      #Length of input sequence
embedding_size = 512   #Size of embedding vector
hidden_size = 75     #Number of neurons in hidden layer
batch_size = 32       #Number of training examples in one batch
vocab_size = 893      #Vocabulary size of input sequence
slot_size = 125       #Vocabulary size of Slots
intent_size = 22      #Vocabulary size of intents
epoch_num = 15        #Number of iterations

#Placeholder to hold input sentences of a batch 
encoder_inputs = tf.placeholder(tf.int32,[input_steps,batch_size],name='encoder_inputs')

#Placeholder to hold actual length of input sentences of a batch 
encoder_inputs_actual_length = tf.placeholder(tf.int32, [batch_size],
                                                   name='encoder_inputs_actual_length')
#Placeholder to hold output given by decoder i.e., predicted slot of a batch
decoder_targets = tf.placeholder(tf.int32, [batch_size, input_steps],
                                      name='decoder_targets')
#Placeholder to hold predicted intent of a batch
intent_targets = tf.placeholder(tf.int32, [batch_size],
                                     name='intent_targets')

#Embeddings
embeddings = tf.Variable(tf.random_uniform([vocab_size, embedding_size],-1.0, 1.0), dtype=tf.float32, name="embedding")
#This embedding matrix will be fed to the encoder as input
encoder_inputs_embedded = tf.nn.embedding_lookup(embeddings, encoder_inputs)
encoder_inputs_embedded
```

```python
#Encoder
from tensorflow.contrib.rnn import LSTMCell, LSTMStateTuple

#Number of cells in Forward RNN 
encoder_f_cell = LSTMCell(hidden_size)
#Number of cells in Backward RNN 
encoder_b_cell = LSTMCell(hidden_size)

'''Feeding embedding matrix of input to the Encoder which is Bidirectional RNN with number of LSTM cells in backward and forward
RNN is same which is 50 here'''
''' Dynamic RNN puts Embedding layer and RNN layers all together'''
(encoder_fw_outputs, encoder_bw_outputs), (encoder_fw_final_state, encoder_bw_final_state) = \
    tf.nn.bidirectional_dynamic_rnn(cell_fw=encoder_f_cell,
                                    cell_bw=encoder_b_cell,
                                    inputs=encoder_inputs_embedded,
                                    sequence_length=encoder_inputs_actual_length,
                                    dtype=tf.float32, time_major=True)

print("encoder_fw_outputs: ", encoder_fw_outputs)
print("encoder_bw_outputs: " ,encoder_bw_outputs)
print("encoder_fw_final_state: " ,encoder_fw_final_state)
print("encoder_bw_final_state: " ,encoder_bw_final_state)

#Concatenating output of Forward and Backward RNN
encoder_outputs = tf.concat((encoder_fw_outputs, encoder_bw_outputs), 2)

#Concatenating final hidden states of Forward and Backward RNN
encoder_final_state_c = tf.concat((encoder_fw_final_state.c, encoder_bw_final_state.c), 1)

#Concatenating final output states of Forward and Backward RNN
encoder_final_state_h = tf.concat((encoder_fw_final_state.h, encoder_bw_final_state.h), 1)

#Concatenating final states
encoder_final_state = LSTMStateTuple(c=encoder_final_state_c,h=encoder_final_state_h)

print("encoder_outputs: ", encoder_outputs)
print("encoder_outputs[0]: ", encoder_outputs[0])
print("encoder_final_state_c: ", encoder_final_state_c)
print("encoder_final_state_h: ", encoder_final_state_h)
print("encoder_final_state: ", encoder_final_state)
```

```python
#Decoder
decoder_lengths = encoder_inputs_actual_length

#Random initializing of weights  
slot_W = tf.Variable(tf.random_uniform([hidden_size * 2, slot_size], -1, 1),
                             dtype=tf.float32, name="slot_W")
intent_W = tf.Variable(tf.random_uniform([hidden_size * 2, intent_size], -1, 1),
                               dtype=tf.float32, name="intent_W")
#Initialization of biases to zeros
slot_b = tf.Variable(tf.zeros([slot_size]), dtype=tf.float32, name="slot_b")
intent_b = tf.Variable(tf.zeros([intent_size]), dtype=tf.float32, name="intent_b")

print("slot_W: ", slot_W)
print("slot_b: ", slot_b)
print("intent_W: ", intent_W)
print("intent_b: ", intent_b)
print("encoder_final_state_h: ", encoder_final_state_h)

# start intent
intent_logits = tf.add(tf.matmul(encoder_final_state_h, intent_W), intent_b)
print("intent_logits: ", intent_logits)
intent = tf.argmax(intent_logits, axis=1)
print("intent: ", intent)

#Multiply with 2 because index of BOS is 2
bos_time_slice = tf.ones([batch_size], dtype=tf.int32, name='BOS') * 2
bos_step_embedded = tf.nn.embedding_lookup(embeddings, bos_time_slice)
pad_step_embedded = tf.zeros([batch_size, hidden_size * 2 + embedding_size],
                             dtype=tf.float32)
print("bos_time_slice: ", bos_time_slice)
print("bos_step_embedded: ", bos_step_embedded)
print("pad_step_embedded: ", pad_step_embedded)
```

```python
def initial_fn():
    #Initially all are false
    initial_elements_finished = (0 >= decoder_lengths)  
    initial_input = tf.concat((bos_step_embedded, encoder_outputs[0]), 1)
    print("initial_elements_finished: ", initial_elements_finished)
    print("initial_input: ", initial_input)
    return initial_elements_finished, initial_input

def sample_fn(time, outputs, state):
    #Select the index that has maximum value as prediction_id
    prediction_id = tf.to_int32(tf.argmax(outputs, axis=1))
    print("prediction_id: ", prediction_id)
    return prediction_id

def next_inputs_fn(time, outputs, state, sample_ids):
    #The output class on the previous time node, gets embedding and enters as the next time node
    pred_embedding = tf.nn.embedding_lookup(embeddings, sample_ids)
    #Concatenating output from previous time node and output of present time node from encoder
    next_input = tf.concat((pred_embedding, encoder_outputs[time]), 1)
    #Tells if time is greater than actual length of inputs. Tensor of batch size.
    elements_finished = (time >= decoder_lengths)  
    #Tells if all actual inputs are finished or not. A boolean scalar.
    all_finished = tf.reduce_all(elements_finished)
    #If all_finished is True then next_inputs is
    next_inputs = tf.cond(all_finished, lambda: pad_step_embedded, lambda: next_input)
    next_state = state
    return elements_finished, next_inputs, next_state

#Base abstract class that allows us to customize sampling
#Helper is where we pass the embeded input. this is a helper instance. 
#This instance should be delivered to the BasicDecoder, which is the actual process of building the decoder model.
my_helper = tf.contrib.seq2seq.CustomHelper(initial_fn, sample_fn, next_inputs_fn)

def decode(helper, scope, reuse=None):
    with tf.variable_scope(scope, reuse=reuse):
        memory = tf.transpose(encoder_outputs, [1, 0, 2])
        print("memory: ",memory)
        #Applying Bahdanau Attention 
        attention_mechanism = tf.contrib.seq2seq.BahdanauAttention(
            num_units=hidden_size, memory=memory,
            memory_sequence_length=encoder_inputs_actual_length)
        print("attention_mechanism: ", attention_mechanism)
        cell = tf.contrib.rnn.LSTMCell(num_units=hidden_size * 2)
        attn_cell = tf.contrib.seq2seq.AttentionWrapper(
            cell, attention_mechanism, attention_layer_size=hidden_size)
        print("attn_cell: ", attn_cell)
        out_cell = tf.contrib.rnn.OutputProjectionWrapper(
            attn_cell, slot_size, reuse=reuse
        )
        print("out_cell: ",out_cell)
        #BasicDecoder builds the decoder model it connects the RNN layers on the decoder side and the input prepared by Helper.
        decoder = tf.contrib.seq2seq.BasicDecoder(
            cell=out_cell, helper=helper,
            initial_state=out_cell.zero_state(
                dtype=tf.float32, batch_size=batch_size))
        print("decoder: ", decoder)
        #dynamic_decode unrolls the decoder model so that actual prediction can be retrieved by BasicDecoder for each time steps.
        final_outputs, final_state, final_sequence_lengths = tf.contrib.seq2seq.dynamic_decode(
            decoder=decoder, output_time_major=True,
            impute_finished=True, maximum_iterations=input_steps
        )
        print("final_outputs: ", final_outputs)
        print("final_state: ", final_state)
        print("final_sequence_length: ",final_sequence_lengths)
        return final_outputs
    
outputs = decode(my_helper, 'decode')

print("outputs: ", outputs)
print("outputs.rnn_output: ", outputs.rnn_output)
print("outputs.sample_id: ", outputs.sample_id)
```

```python
decoder_prediction = outputs.sample_id
decoder_max_steps, decoder_batch_size, decoder_dim = tf.unstack(tf.shape(outputs.rnn_output))
decoder_targets_time_majored = tf.transpose(decoder_targets, [1, 0])
decoder_targets_true_length = decoder_targets_time_majored[:decoder_max_steps]
print("decoder_prediction: ", decoder_prediction)
print("decoder_max_steps: ", decoder_max_steps)
print("decoder_batch_size: ", decoder_batch_size)
print("decoder_targets_time_majored: ", decoder_targets_time_majored)
print("decoder_targets_true_length: ", decoder_targets_true_length)
```

```python
# Define mask so that padding does not count towards loss
mask = tf.to_float(tf.not_equal(decoder_targets_true_length, 0))

# Defining the loss of the slot label
loss_slot = tf.contrib.seq2seq.sequence_loss(
    outputs.rnn_output, decoder_targets_true_length, weights=mask)

# Defining loss of intent classification
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(
    labels=tf.one_hot(intent_targets, depth=intent_size, dtype=tf.float32),
    logits=intent_logits)
loss_intent = tf.reduce_mean(cross_entropy)
```

```python
# Metrics Functions
import numpy.ma as ma
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

loss = loss_slot + loss_intent
optimizer = tf.train.AdamOptimizer(name="a_optimizer")
grads, vars = zip(*optimizer.compute_gradients(loss))
gradients, _ = tf.clip_by_global_norm(grads, 5)  # clip gradients
train_op = optimizer.apply_gradients(zip(grads, vars))

def step(sess, mode, train_batch):
    #perform for each batch
    unziped = list(zip(*train_batch))
    if mode == 'train':
        output_feeds = [train_op, loss, decoder_prediction,
                        intent]
        feed_dict = {encoder_inputs: np.transpose(unziped[0], [1, 0]),
                     encoder_inputs_actual_length: unziped[1],
                     decoder_targets: unziped[2],
                     intent_targets: unziped[3]}
    if mode in ['test']:
        output_feeds = [decoder_prediction, intent]
        feed_dict = {encoder_inputs: np.transpose(unziped[0], [1, 0]),
                     encoder_inputs_actual_length: unziped[1]}
    results = sess.run(output_feeds, feed_dict=feed_dict)
    return results

def accuracy_score(true_data, pred_data, true_length=None):
    true_data = np.array(true_data)
    pred_data = np.array(pred_data)
    assert true_data.shape == pred_data.shape
    if true_length is not None:
        val_num = np.sum(true_length)
        assert val_num != 0
        res = 0
        for i in range(true_data.shape[0]):
            res += np.sum(true_data[i, :true_length[i]] == pred_data[i, :true_length[i]])
    else:
        val_num = np.prod(true_data.shape)
        assert val_num != 0
        res = np.sum(true_data == pred_data)
    res /= float(val_num)
    return res

def get_data_from_sequence_batch(true_batch, pred_batch, padding_token):
    true_ma = ma.masked_equal(true_batch, padding_token)
    pred_ma = ma.masked_array(pred_batch, true_ma.mask)
    true_ma = true_ma.flatten()
    pred_ma = pred_ma.flatten()
    true_ma = true_ma[~true_ma.mask]
    pred_ma = pred_ma[~pred_ma.mask]
    return true_ma, pred_ma

def f1_for_sequence_batch(true_batch, pred_batch, average="micro", padding_token=0):
    true, pred = get_data_from_sequence_batch(true_batch, pred_batch, padding_token)
    labels = list(set(true))
    return f1_score(true, pred, labels=labels, average=average)

def accuracy_for_sequence_batch(true_batch, pred_batch, padding_token=0):
    true, pred = get_data_from_sequence_batch(true_batch, pred_batch, padding_token)
    return accuracy_score(true, pred)

def getBatch(batch_size, train_data):
    random.shuffle(train_data)
    sindex = 0
    eindex = batch_size
    while eindex < len(train_data):
        batch = train_data[sindex:eindex]
        temp = eindex
        eindex = eindex + batch_size
        sindex = temp
        yield batch
        
def getBatch2(batch_size, train_data):
    sindex = 0
    eindex = batch_size
    while eindex < len(train_data):
        batch = train_data[sindex:eindex]
        temp = eindex
        eindex = eindex + batch_size
        sindex = temp
        yield batch
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

epochs = []
F1_Score = []

for epoch in range(epoch_num):
    epochs.append(epoch)
    mean_loss = 0.0
    train_loss = 0.0
    for i, batch in enumerate(getBatch(batch_size, train_data)):
        _, loss_v, decoder_prediction_v, intent_v = step(sess, "train", batch)
        mean_loss += loss_v
        train_loss += loss_v
        if i % 30 == 0:
            if i > 0:
                mean_loss = mean_loss / 30.0
            print('Average train loss at epoch %d, step %d: %f' % (epoch, i, mean_loss))
            mean_loss = 0
    train_loss /= (i + 1)
    print("[Epoch {}] Average train loss: {}".format(epoch, train_loss))
    
    # One epoch per training, test once
    pred_slots = []
    for j, batch in enumerate(getBatch(batch_size, val_data)):
        decoder_prediction_v, intent_v = step(sess, "test", batch)
        decoder_prediction_v = np.transpose(decoder_prediction_v, [1, 0])
        if j == 0:
            index = random.choice(range(len(batch)))
            print("Input Sentence        : ", index_seq2word(batch[index][0], index2word))
            print("Slot Truth            : ", index_seq2slot(batch[index][2], index2tag))
            print("Slot Prediction       : ", index_seq2slot(decoder_prediction_v[index], index2tag))
            print("Intent Truth          : ", index2intent[batch[index][3]])
            print("Intent Prediction     : ", index2intent[intent_v[index]])
        slot_pred_length = list(np.shape(decoder_prediction_v))[1]
        pred_padded = np.lib.pad(decoder_prediction_v, ((0, 0), (0, input_steps-slot_pred_length)),
                                 mode="constant", constant_values=0)
        pred_slots.append(pred_padded)
        true_slot = np.array((list(zip(*batch))[2]))
        true_length = np.array((list(zip(*batch))[1]))
        true_slot = true_slot[:, :slot_pred_length]
        slot_acc = accuracy_score(true_slot, decoder_prediction_v, true_length)
        intent_acc = accuracy_score(list(zip(*batch))[3], intent_v)
        print("slot accuracy: {}, intent accuracy: {}".format(slot_acc, intent_acc))
    
    pred_slots_a = np.vstack(pred_slots)
    true_slots_a = np.array(list(zip(*val_data))[2])[:pred_slots_a.shape[0]]
    print("Test F1 score for epoch {}: {}".format(epoch, f1_for_sequence_batch(true_slots_a, pred_slots_a)))
    F1_Score.append(f1_for_sequence_batch(true_slots_a, pred_slots_a))
plt.plot(epochs,F1_Score)
plt.scatter(epochs,F1_Score)
plt.show()
```
